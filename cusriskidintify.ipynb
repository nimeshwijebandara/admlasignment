```python
import streamlit as st
import joblib
import pandas as pd


model = joblib.load("credit_risk_model.pkl")

st.title("People's Leasing & Finance PLC")
st.subheader("Credit Risk Prediction System")


age = st.number_input("Age", 18, 70, 30)
income = st.number_input("Monthly Income", 5000, 500000, 50000)
loan = st.number_input("Loan Amount", 10000, 5000000, 100000)
liability = st.number_input("Existing Liabilities", 0, 3000000, 0)
history = st.number_input("Credit History Length", 0, 30, 5)

if st.button("Predict Risk"):

    data = pd.DataFrame([[
        age,
        income,
        loan,
        liability,
        history
    ]], columns=[
        "Age",
        "Montly income",
        "Loan Amount",
        "Exsisting Liabilities",
        "Credit History Length"
    ])

    prediction = model.predict(data)

    if prediction[0] == 1:
        st.error("High Risk Customer")
    else:
        st.success("Low Risk Customer")
```
